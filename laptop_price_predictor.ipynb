```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')

new = df["Memory"].str.split("+", n=1, expand=True)

df["first"] = new[0]
df["first"] = df["first"].str.strip()

df["second"] = new[1]
df["second"].fillna("0", inplace=True)

# --- Process the first storage component ---
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Clean the string, keeping only digits
df['first'] = df['first'].str.replace(r'\D', '')

# --- ROBUST FIX for 'first' column ---
# Convert to numeric, turning errors into NaN (Not a Number)
df['first'] = pd.to_numeric(df['first'], errors='coerce')
# Fill any NaN values with 0
df['first'].fillna(0, inplace=True)


# --- Process the second storage component ---
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Clean the string, keeping only digits
df['second'] = df['second'].str.replace(r'\D', '')

# --- ROBUST FIX for 'second' column ---
# Convert to numeric, turning errors into NaN
df['second'] = pd.to_numeric(df['second'], errors='coerce')
# Fill any NaN values with 0
df['second'].fillna(0, inplace=True)


# --- Convert to integers and calculate totals ---
df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# --- Drop temporary columns ---
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV, cross_val_score
from sklearn.inspection import permutation_importance
import time
```

```python
# using hypertunning improving performance of the model 
cat_idxs = [0,1,10,13,14]

# safer ColumnTransformer: handle_unknown, use sparse_output=True when many categories
pre = ColumnTransformer([
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), cat_idxs)
], remainder='passthrough')   # passthrough keeps numeric columns as-is

rf = RandomForestRegressor(
    n_estimators=200,     # baseline higher than 100
    random_state=3,
    n_jobs=-1,
    bootstrap=True
)

pipe_rand = Pipeline([
    ('pre', pre),
    ('rf', rf)
])

# --- Randomized search space (sensible ranges) ---
param_dist = {
    'rf__n_estimators': [100, 200, 400, 800],
    'rf__max_depth': [6, 10, 15, 25, None],
    'rf__max_features': [0.3, 0.5, 0.7, 0.9, 'sqrt', 'log2'],
    'rf__min_samples_split': [2, 4, 8, 12],
    'rf__min_samples_leaf': [1, 2, 4, 8],
    'rf__bootstrap': [True, False]
}

rnd = RandomizedSearchCV(
    pipe_rand,
    param_distributions=param_dist,
    n_iter=40,           # adjust up to 100 if you have time
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

start = time.time()
rnd.fit(X_train, y_train)
print("RandomizedSearch done in {:.1f}s".format(time.time()-start))
print("Best params:", rnd.best_params_)
print("Best CV R2 (mean):", rnd.best_score_)

# Evaluate best estimator on test set
best = rnd.best_estimator_
y_pred = best.predict(X_test)
print("Test R2:", r2_score(y_test, y_pred))
print("Test MAE:", mean_absolute_error(y_test, y_pred))

# Cross-val on the best estimator (good to confirm stability)
cv_scores = cross_val_score(best, X_train, y_train, cv=5, scoring='r2', n_jobs=-1)
print("Train CV R2 mean/std:", cv_scores.mean(), cv_scores.std())

# --- Feature importance: map back to original feature names after OHE ---
# Build feature names: needs training data to get categories from OneHotEncoder
pre_fit = best.named_steps['pre']
# if X_train is a DataFrame, use its columns; else create generic names
if hasattr(X_train, 'columns'):
    col_names = []
    # get transformer output feature names
    try:
        # for OneHotEncoder inside ColumnTransformer
        ohe = pre_fit.named_transformers_['cat']
        ohe_feature_names = ohe.get_feature_names_out([str(i) for i in cat_idxs])
    except Exception:
        # fallback: use get_feature_names_out with original df columns slice
        ohe_feature_names = pre_fit.named_transformers_['cat'].get_feature_names_out()
    # remainder columns (passthrough) ordering: columns not in cat_idxs
    remainder_cols = [c for i, c in enumerate(X_train.columns) if i not in cat_idxs]
    col_names = list(ohe_feature_names) + remainder_cols
else:
    # fallback generic names
    # We can't reliably map without a DataFrame; create indices
    transformed = pre_fit.transform(X_train[:2])  # small transform to get shape
    n_features = transformed.shape[1]
    col_names = [f'F{i}' for i in range(n_features)]

# get importances
importances = best.named_steps['rf'].feature_importances_
imp_df = pd.DataFrame({'feature': col_names, 'importance': importances})
imp_df = imp_df.sort_values('importance', ascending=False).reset_index(drop=True)
print(imp_df.head(30))

# Permutation importance (more reliable)
perm = permutation_importance(best, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)
perm_idx = np.argsort(perm.importances_mean)[::-1]
print("Top permutation importances:")
for i in perm_idx[:20]:
    print(col_names[i], perm.importances_mean[i].round(4), "+/-", perm.importances_std[i].round(4))
```

# improved SVM with grid search

```python
cat_idxs = [0,1,10,13,14]    # your categorical indices

# if X_train is a pandas DataFrame, build lists by dtype
if hasattr(X_train, "dtypes"):
    cat_names = [X_train.columns[i] for i in cat_idxs]
    num_names = [c for c in X_train.columns if c not in cat_names]
    transformer = ColumnTransformer([
        ('num', StandardScaler(), num_names),
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), cat_names)
    ])
else:
    # numpy array: numeric indices = all except cat_idxs
    all_idxs = list(range(X_train.shape[1]))
    num_idxs = [i for i in all_idxs if i not in cat_idxs]
    transformer = ColumnTransformer([
        ('num', StandardScaler(), num_idxs),
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), cat_idxs)
    ])

pipe = Pipeline([
    ('pre', transformer),
    ('svr', SVR(kernel='rbf'))
])

# Grid search for core hyperparams (smaller grid to start)
param_grid = {
    'svr__C': [10, 100, 1000, 10000],
    'svr__gamma': ['scale', 0.01, 0.001, 0.0001],
    'svr__epsilon': [0.01, 0.05, 0.1, 0.2]
}

grid = GridSearchCV(pipe, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)

print("BEST", grid.best_params_, "CV R2:", grid.best_score_)
y_pred = grid.predict(X_test)
print("Test R2:", r2_score(y_test, y_pred))
print("Test MAE:", mean_absolute_error(y_test, y_pred))
```
